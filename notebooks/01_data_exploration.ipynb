# 01 – Data Exploration

In this notebook we perform an initial exploration of the Lending Club dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed sample data for quick exploration
df = pd.read_csv('../data/sample_data/lending_club_sample.csv')

# Display first five rows
df.head()
```

```
# Summary statistics
df.describe(include='all')
```

The dataset contains thousands of loans with dozens of variables.  Missing values and high cardinality categorical features are handled in the preprocessing step.

```
# Correlation heatmap on numeric features
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=['float', 'int']).corr(), cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
